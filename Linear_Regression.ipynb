### Gradient descent ->
 is used to minimize the cost functionof a model by adjusting its parameters.

 Basic idea of the gradient descent is to compute the (partial derivatives)gradient of the cost function with respect to each parameters in the model.
# New Section
 Some steps involved in gradient descents are
 - 1 -> Initialize parameters : Start by initializing the parameters of the model with some initila values.
 - 2-> compute the cost : We will evaluate the cost function by f3eding the inout data to the model and comapring th predcited output with the actual output.
 - 3-> Compute the Gradient -> Calculate the partial derivative of the cost function with respect to each parameter we will pass in our model. Gradient indicates the steepest ascent.
 - 4-> Update parameters -> Adjust the parameters values by taking step in the opposite  direction of the gradient. This step is determined by a learning rate, whihc controls the size of the update at each iteration.
 - 5-> Repeat -> Repeat step 2 to 4 until convergence or a predefiend number of iterations.
 Convergence occurs when the cost function reaches a minimum or when the updates to the parameters become very small.
